```python
import csv

# Path to your original CSV file
input_csv_file_path = '../github-pipelines/length1_90/test_0_dirty.csv'

# Path to the new CSV file with bigint-compatible format
output_csv_file_path = '../github-pipelines/length1_90/test_0_clean.csv'

# List of column names that need to be converted from scientific notation
columns_to_convert = ['tweet_id', 'in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id']

# Read the original CSV and write to a new one with conversions
with open(input_csv_file_path, newline='', encoding='utf-8') as infile, \
     open(output_csv_file_path, 'w', newline='', encoding='utf-8') as outfile:

    reader = csv.DictReader(infile)
    writer = csv.DictWriter(outfile, fieldnames=reader.fieldnames)

    writer.writeheader()

    for row in reader:
        for column in columns_to_convert:
            # Convert scientific notation to bigint-compatible format
            if row[column] and row[column] != '':
                row[column] = str(int(float(row[column])))
            else:
             #iter1 Error in the previous response: Error: ERROR:  invalid input syntax for type double precision:
             #CONTEXT:  COPY Source1_90_0, line 2, column in_reply_to_status_id: 
                # Handle empty strings - set to None or a default value
                row[column] = None

        writer.writerow(row)

print("CSV preprocessing complete.")
```
